# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time


class Kumanda:
    def __init__(self):
        self.tv_durum = "Kapalı"
        self.ses = 10
        self.kanallar = ["Kanal 1", "Kanal 2", "Kanal 3", "Kanal 4"]
        self.aktif_kanal = None

    def tv_ac(self):
        if self.tv_durum == "Açık":
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            time.sleep(1)
            self.tv_durum = "Açık"
            self.aktif_kanal = self.kanallar[0]
            print("Televizyon açıldı.")

    def tv_kapat(self):
        if self.tv_durum == "Kapalı":
            print("Televizyon zaten kapalı.")
        else:
            print("Televizyon kapanıyor...")
            time.sleep(1)
            self.tv_durum = "Kapalı"
            self.aktif_kanal = None
            print("Televizyon kapandı.")

    def ses_ayarla(self, yeni_ses):
        if 0 <= yeni_ses <= 100:
            self.ses = yeni_ses
            print(f"Ses seviyesi {self.ses} olarak ayarlandı.")
        else:
            print("Ses seviyesi 0 ile 100 arasında olmalıdır.")

    def kanal_ekle(self, kanal_adi):
        if kanal_adi not in self.kanallar:
            self.kanallar.append(kanal_adi)
            print(f"{kanal_adi} kanalı eklendi.")
        else:
            print(f"{kanal_adi} kanalı zaten mevcut.")

    def kanal_sayisi(self):
        print(f"Toplam {len(self.kanallar)} kanal var.")

    def rastgele_kanal(self):
        if self.tv_durum == "Açık":
            self.aktif_kanal = random.choice(self.kanallar)
            print(f"Rastgele bir kanala geçildi: {self.aktif_kanal}")
        else:
            print("Televizyon kapalı. Lütfen önce televizyonu açın.")

    def tv_bilgileri(self):
        print(f"Televizyon Durumu: {self.tv_durum}")
        print(f"Ses Seviyesi: {self.ses}")
        print(f"Mevcut Kanal: {self.aktif_kanal}")
        print(f"Kanallar: {', '.join(self.kanallar)}")

kumanda = Kumanda()

while True:
    print("\n1. Tv Aç\n2. Tv Kapat\n3. Ses Ayarları\n4. Kanal Ekle\n5. Kanal Sayısını Öğrenme\n6. Rastgele Kanala Geçme\n7. Televizyon Bilgileri\nq. Çıkış")
    secim = input("Bir işlem seçin: ")

    if secim == "1":
        kumanda.tv_ac()
    elif secim == "2":
        kumanda.tv_kapat()
    elif secim == "3":
        yeni_ses = int(input("Yeni ses seviyesini girin (0-100): "))
        kumanda.ses_ayarla(yeni_ses)
    elif secim == "4":
        kanal_adi = input("Eklemek istediğiniz kanal adını girin: ")
        kumanda.kanal_ekle(kanal_adi)
    elif secim == "5":
        kumanda.kanal_sayisi()
    elif secim == "6":
        kumanda.rastgele_kanal()
    elif secim == "7":
        kumanda.tv_bilgileri()
    elif secim == "q":
        print("Programdan çıkılıyor...")
        break
    else:
        print("Geçersiz bir seçim yaptınız.")
```
