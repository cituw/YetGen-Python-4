# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def vki_hesapla(boy, kilo):
    return kilo / (boy / 100) ** 2

def vki_siniflandir(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki < 25:
        return "Normal kilolu"
    elif 25 <= vki < 30:
        return "Fazla kilolu"
    elif 30 <= vki < 40:
        return "Obez"
    else:
        return "İleri derecede obez"

def verileri_dosyaya_yaz(dosya_adi, veriler):
    with open(dosya_adi, "a", encoding="utf-8") as dosya:
        for veri in veriler:
            dosya.write(f"{veri['isim']}, {veri['yas']}, {veri['boy']}, {veri['kilo']}, {veri['vki']:.2f}, {veri['sinif']}\n")

def main():
    dosya_adi = "vki_sonuclari.txt"
    giris_sayisi = int(input("Kaç giriş yapacaksınız? "))
    veriler = []

    for _ in range(giris_sayisi):
        isim = input("İsim-Soyisim: ")
        yas = int(input("Yaş: "))
        boy = float(input("Boy (cm): "))
        kilo = float(input("Kilo (kg): "))

        vki = vki_hesapla(boy, kilo)
        sinif = vki_siniflandir(vki)

        veriler.append({
            "isim": isim,
            "yas": yas,
            "boy": boy,
            "kilo": kilo,
            "vki": vki,
            "sinif": sinif
        })

        print(f"{isim} için VKİ: {vki:.2f}, Sınıf: {sinif}")

    verileri_dosyaya_yaz(dosya_adi, veriler)
    print(f"Tüm veriler '{dosya_adi}' dosyasına kaydedildi.")

main()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime

MENU = {
    1: ("Çilekli Smoothie", 30),
    2: ("Muzlu Smoothie", 30),
    3: ("Yeşil Smoothie", 35),
    4: ("Protein Smoothie", 40)
}

def menu_goruntule():
    print("YetBar Menü:")
    for key, (urun, fiyat) in MENU.items():
        print(f"{key}. {urun}: {fiyat} TL")

def satin_al():
    toplam_tutar = 0
    secimler = []

    while True:
        menu_goruntule()
        secim = input("Bir ürün seçin (çıkmak için 'q' tuşlayın): ").strip()

        if secim.lower() == 'q':
            break
        elif secim.isdigit() and int(secim) in MENU:
            urun, fiyat = MENU[int(secim)]
            secimler.append((urun, fiyat))
            toplam_tutar += fiyat
            print(f"{urun} seçildi. Toplam tutar: {toplam_tutar} TL")
        else:
            print("Geçersiz seçim. Lütfen menüden bir ürün seçin.")

    if secimler:
        tarih = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open("smoothie_satislari.txt", "a", encoding="utf-8") as dosya:
            for urun, fiyat in secimler:
                dosya.write(f"{tarih}, {urun}, {fiyat} TL\n")
        print(f"Satın alma işlemi tamamlandı. Toplam tutar: {toplam_tutar} TL")
        print("Satış bilgileri 'smoothie_satislari.txt' dosyasına kaydedildi.")
    else:
        print("Herhangi bir ürün seçilmedi.")

def main():
    while True:
        print("\nYetBar - Smoothie Bank")
        print("1. Menüyü Görüntüle")
        print("2. Smoothie Satın Al")
        print("3. Çıkış Yap")
        secim = input("Bir işlem seçin: ").strip()

        if secim == "1":
            menu_goruntule()
        elif secim == "2":
            satin_al()
        elif secim == "3":
            print("Çıkış yapılıyor. İyi günler!")
            break
        else:
            print("Geçersiz seçim. Lütfen tekrar deneyin.")
main()
```
